```python
from matplotlib import pyplot as plt
with open("temperatures_evolution.txt") as f:
    s = f.readline()
# pop last ','   
s = '[' + s[:-1] + ']'

l = eval(s)

plt.plot(l, label="λ=0.995, T_0=1000, i=100, T_x+=100")
plt.legend()
plt.title("Tetravex 5x5")
plt.xlabel("Nombre d'iterations")
plt.ylabel("Température")
plt.show()
```
